```python
dna = "ATGCGATACGCTTGA"

A = dna.count("A")
T = dna.count("T")
G = dna.count("G")
C = dna.count("C")

print("A:" , A)
print("T:" , T)
print("G:" , G)
print("C:" , C)

print("DNA Length:" , len(dna))

gc = (G+C)/len(dna) * 300
print("GC Content:" , gc)

import matplotlib.pyplot as plt

labels = ["A" , "T" , "G" , "C"]
counts = [A , T , G , C]

plt.bar(labels, counts)

plt.title("Nucleotide Distribution")
plt.xlabel("Bases")
plt.ylabel("Count")
plt.show()
```
